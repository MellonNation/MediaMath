```python
import json
from tqdm.notebook import tqdm
from argparse import Namespace
import os
import re
```

```python
args = Namespace(
jsonFile = "context_{}.json",
skiprow = 100000
)
```

```python
listfiles = os.listdir()
listfiles = [_ for _ in listfiles if re.match("context_\d",_)]
listfiles.sort()
for listfile in listfiles:
    os.remove(listfile)
```

```python
with open("context.json","r") as f:
    readlines = json.load(f)
dicts = []
for i in tqdm(range(len(readlines)), desc='Generate rows'):
    dictstr = readlines[i]
    dicts.append(json.loads(dictstr))
```

```python
mydict = dicts[0]
for i in tqdm(range(len(dicts)), desc='Generate rows'):
    if not i:
        continue
    mydict.update(dicts[i])
    if i%args.skiprow==0:
        with open(args.jsonFile.format(i/args.skiprow),"w") as f:
            json.dump(mydict,f)
with open(args.jsonFile.format("all"),"w") as f:
    json.dump(mydict,f)
```
